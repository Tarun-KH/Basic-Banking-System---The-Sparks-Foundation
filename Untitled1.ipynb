```python
import matplotlib.pyplot as plt
latitude_points = [10.0, 10.2, 10.5, 10.8, 11.0, 11.2, 11.5]
plt.figure(figsize=(8, 6))
plt.plot(range(len(latitude_points)), latitude_points, marker='o', linestyle='-')
plt.xlabel('Data Points')
plt.ylabel('Latitude')
plt.title('Latitude Points of a City')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt


countries = ['India', 'England', 'USA']


population_2023 = [150000000, 90000000, 120000000]


plt.figure(figsize=(10, 6))
plt.bar(countries, population_2023, color=['blue', 'green', 'red'])

plt.xlabel('Countries in world ')
plt.ylabel('Population (2023)')
plt.title('Population of Three Different Countries in 2023')


plt.show()
```

```python
import matplotlib.pyplot as plt
factors = ['Consumption', 'Investment', 'Government Spending', 'Net Exports']
percentages = [60, 20, 18, 5]
explode = (0.1, 0, 0.2, 0)
plt.figure(figsize=(8, 8))
plt.pie(percentages, labels=factors, autopct='%1.1f%%', startangle=140, explode=explode)
plt.title('Factors Contributing to GDP')
plt.axis('equal')
plt.show()
```
